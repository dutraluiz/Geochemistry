# 2. Descritive statistics

```python
stats_list = []

for col in element_list:
    mean_val = df[col].mean()
    std_val = df[col].std()
    
    # CV (%) = (desvio padrão / média) * 100
    cv_val = (std_val / mean_val) * 100 if mean_val != 0 else 0
    
    stats_list.append({'Elemento': col, 'Média': mean_val, 'Mediana': df[col].median(),
                       'Mínimo': df[col].min(), 'Máximo': df[col].max(), 'Desvio Padrão': std_val,
                       'CV (%)': cv_val})

# Criar DataFrame de resultados / Create results DataFrame
df_resumo = pd.DataFrame(stats_list)

# Exibir a tabela / Display the table
print('Estatística Descritiva:')
display(df_resumo.round(3))

# Salvar df_resumo / Save df_resumo
df_resumo.to_excel('Data/' + file_name + '_DESC_STATISTICS.xlsx', index=False)
```

# 4. Variability Analysis

```python
# 1. Cálculo das estatísticas
cv_stats = pd.DataFrame({
    'Média': df[element_list].mean(), 'Desvio Padrão': df[element_list].std()})

# 2. Cálculo do CV em porcentagem
cv_stats['CV (%)'] = (cv_stats['Desvio Padrão'] / cv_stats['Média']) * 100

# 3. Função de classificação conforme seu critério
def classificar_homogeneidade(cv):
    if cv < 20:
        return 'Baixa'
    elif 20 <= cv <= 50:
        return 'Moderada'
    else:
        return 'Alta'

# 4. Aplicação da classificação
cv_stats['Variabilidade'] = cv_stats['CV (%)'].apply(classificar_homogeneidade)

# Ordenar para facilitar a triagem (Etapa 2)
cv_stats = cv_stats.sort_values('CV (%)', ascending=False)

print(cv_stats.round(1))

# Opcional: Salvar em CSV para o relatório
cv_stats.to_excel('Data/' + file_name + '_CV.xlsx')
```

# 7. CLR transformation

```python
import numpy as np
from scipy.stats import gmean

df_clr = df[element_list]

# 2. Função para o Centered Log-Ratio (CLR)
def clr_transform(df):
    # Calcula a média geométrica por linha (amostra)
    geom_mean = gmean(df, axis=1)
    # Aplica o logaritmo da razão entre o valor e a média geométrica
    return np.log(df.divide(geom_mean, axis=0))

# 3. Execução
df_transformed = clr_transform(df_clr)

# 4. Remove os sufixos '_ppm' e '_pct' dos nomes das colunas transformadas
# O regex '(_ppm|_pct)$' procura por qualquer um dos dois no final da string
df_transformed.columns = df_transformed.columns.str.replace(r'(_ppm|_pct)$', '', regex=True)

# 5. Adiciona o sufixo '_CLR' aos nomes limpos
df_transformed_labeled = df_transformed.add_suffix('_CLR')

# 6. Concatenar (juntar) com o DataFrame original lateralmente
# O Pandas utiliza o índice para garantir que a linha da amostra coincida
df_final = pd.concat([df, df_transformed_labeled], axis=1)

# 7. Salvar em Excel (.xlsx)
# O engine 'openpyxl' é o padrão para escrita de ficheiros Excel modernos
df_final.to_excel('Data/' + file_name + '_CLR.xlsx', index=False)
```

# 8. Interpretation

A interpretação das correlação requer o conhecimento do significado geoquímico e geológico a ocorrência dos elementos. <br>

O Zn apresenta correlação moderada com Co (ρ = 0,70), Mn (ρ = 0,63), S (ρ = 0,61), Te (ρ = 0,54) e correlação negativa com U (ρ = -0,57). O Zn é um elemento calcófilo de alta mobilidade geoquímica, comportamento semelhante ao do Co. A associação com Mn e S pode estar relacionada a processos de lixiviação geoquímica de rochas félsicas e máficas, bem como à adsorção em óxidos de manganês e à presença de fases sulfetadas. As correlações com Te e U sugerem a participação de fluidos hidrotermais (Te) e possível influência de fontes crustais graníticas ou fluidos oxidados (U). Assim, os dados indicam que o teor de Zn está associado a uma complexa interação geoquímica envolvendo diferentes litologias e a sobreposição de processos hidrotermais.

**Tabela 1**. Resumo das características geoquímicas de cobalto, zinco, telúrio e urânio.
| Elemento | Afinidade / Caráter (Goldschmidt) | Local de Concentração (Magmático/Crustal) | Mobilidade na Superfície |
|---------|----------------------------------|-----------------------------------------|------------------------|
| Zn (Zinco) | Calcófilo | Rochas máficas e intermediárias / sulfetos | Alta |
| Co (Cobalto) | Siderófilo / Calcófilo | Manto e rochas máficas | Moderada |
| Mn (Manganês) | Litófilo | Manto e Crosta (substitui Fe e Mg em silicatos e óxidos) | Moderada (forma nódulos e crostas de óxidos) |

| Te (Telúrio) | Calcófilo | Veios hidrotermais / sulfetos e teluretos (raro) | Baixa |
| U (Urânio) | Litófilo | Crosta continental (granitos) | Moderada (solúvel em meios oxidados) ||



# 8. Fractal analysis

## 8.1 Individual plot

```python
import numpy as np
import matplotlib.pyplot as plt

def analise_fractal_ca(df, elemento):
    # 1. Preparação dos dados: ordenar e calcular área acumulada
    data = np.sort(df[elemento].dropna().values)[::-1]
    area_acumulada = np.arange(1, len(data) + 1)
    
    # 2. Plotagem Log-Log
    plt.figure(figsize=(8, 6))
    plt.loglog(data, area_acumulada, 'o', markersize=4, color='darkblue', alpha=0.6)
    
    # Estética do gráfico
    plt.title(f'Modelo Fractal C-A: {elemento.replace('_CLR', '')}', fontsize=14, weight='bold')
    plt.xlabel('Concentração (Log)', fontsize=12, weight='bold')
    plt.ylabel('Área Acumulada A(C) (Log)', fontsize=12, weight='bold')
    plt.grid(True, which='both', ls='--', alpha=0.5)
    
    # 3. Salvando em alta resolução
    nome_arquivo = f'fractal_ca_{elemento}.tif'
    plt.savefig('Figure/FA individual/'+nome_arquivo, dpi=600, format='tif', bbox_inches='tight')
    plt.show()
    
    print(f'Gráfico fractal salvo como: {nome_arquivo}')

# Exemplo de uso para o primeiro elemento da sua lista CLR
analise_fractal_ca(df_final, element_clr[0])
```

## 8.2 Subplots

```python
import numpy as np
import matplotlib.pyplot as plt

# Configurações de grade
n_elementos = len(element_clr)
n_cols = 5
n_rows = int(np.ceil(n_elementos / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 4 * n_rows))
axes = axes.flatten() # Transforma a matriz de eixos em uma lista linear

for i, elemento in enumerate(element_clr):
    # 1. Preparação dos dados
    data = np.sort(df_final[elemento].dropna().values)[::-1]
    area_acumulada = np.arange(1, len(data) + 1)
    
    # 2. Plotagem no subplot correspondente
    axes[i].loglog(data, area_acumulada, 'o', markersize=2, color='darkblue', alpha=0.5)
    
    # Estética do Subplot
    axes[i].set_title(elemento.replace('_CLR', ''), fontsize=12, weight='bold')
    axes[i].grid(True, which='both', ls='--', alpha=0.4)
    
    # Adiciona labels apenas nas bordas para não poluir
    if i >= (n_rows - 1) * n_cols:
        axes[i].set_xlabel('Conc. (Log)', fontsize=10)
    if i % n_cols == 0:
        axes[i].set_ylabel('Área Acum. (Log)', fontsize=10)

# 3. Remover eixos vazios (caso tenha menos de 35 elementos na lista final)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()

# 4. Salvar o painel completo em 600 DPI
plt.savefig('Figure/6-Fractal_analysis.tif', dpi=600, format='tif', bbox_inches='tight')
plt.show()
```
